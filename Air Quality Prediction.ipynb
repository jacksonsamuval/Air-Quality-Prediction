## Null Value Handling

```python
# in n_hos_beds, we have less than 10% of data is missing
df_Air_Predict.dropna(inplace=True)
```

```python
df_target=df_Air_Predict['Pollutant_Min']
df_feature=df_Air_Predict.drop('Pollutant_Min', axis=1)
```

```python
# segregate the data based on the datatype 
df_num=df_Air_Predict.select_dtypes(include='number')
df_cat=df_Air_Predict.select_dtypes(include='object')
```

### Perform Encoding on Categorical Variable

```python
# use Dummy Encoding
df_encoded=pd.get_dummies(df_cat)

# reset the index
df_encoded=df_encoded.reset_index()
```

```python
# show the encoded data
df_encoded.head()
```

# Perform scaling on Numerical data

```python
# import the standard scalar
from sklearn.preprocessing import StandardScaler

#initiate the standard scalar
ss=StandardScaler()

# scaled the value present in numerical dataframe
scaled=ss.fit_transform(df_num)

# numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)

# reset the index
scaled_df=scaled_df.reset_index()
```

```python
scaled
```

```python
# show the data 
scaled_df.head()
```

### Perform the Concatenate Operation

```python
# concat encoded variables and scaled variables
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
# show the data
df_independent.head()
```

## Split the Data for Training and Testing

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test=train_test_split(df_independent,
                                              df_target,
                                              test_size=0.01)
```

# Apply Linear Regression

```python
# import the model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# initiate the model
LR=LinearRegression()
```

```python
# train the Linear Regression model
model_lr=LR.fit(x_train,y_train)
```
